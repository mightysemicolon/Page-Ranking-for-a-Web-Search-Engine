```python
import numpy as np
import pandas as pd
from numpy.linalg import norm
import networkx as nx
import matplotlib.pyplot as plt
```

# reading data

```python
data = pd.read_csv("sklearn-anatomy.csv")
data
```

# droping unnecessary columns

```python
data.drop(["page_title", "page_length", "link_text", "link_href"], axis=1, inplace=True)
```

# creating adjacency matrix

```python
df = pd.crosstab(data.page_url, data.link_url)
df
```

```python
idx = df.columns.union(df.index)
df = df.reindex(index = idx, columns=idx, fill_value=0)
Adjacency_Matrix = df.to_numpy(dtype=np.float64)
df
```

# visualizing

```python
graph = nx.from_numpy_array(Adjacency_Matrix)
fig, ax = plt.subplots(figsize=(30, 30))
nx.draw(graph)
```

# creating Q

```python
def dev(val):
    return 1 if val == 0 else val
```

```python
for i in range(Adjacency_Matrix.shape[0]):
   Adjacency_Matrix[i, :] = Adjacency_Matrix[i, :] / dev(Adjacency_Matrix[i, :].sum())
Q = Adjacency_Matrix.transpose()
```

# page rank alg

```python
def page_rank(adjacency_matrix, max_iter=1000, epsilon=1e-16, alpha=0.6):

    initial_pagerank = np.random.rand(adjacency_matrix.shape[0]) # r(0)
    v = np.ones(adjacency_matrix.shape[0]) / adjacency_matrix.shape[0] # 1/n * e

    prev_iteration_pagerank = initial_pagerank # r(0)

    y_hat_initial = alpha * np.dot(adjacency_matrix, initial_pagerank) # a * Q * r(0)
    y_initial = y_hat_initial + (1 - norm(y_hat_initial, ord=1)) * v # y = a * Q * r(0) + (1 - norm1(a * Q * r(0))) * 1 / n * e
    next_iteration_pagerank = y_initial / norm(y_initial, ord=1) # r(1) = y / norm1(y)
    iter = 1


    while iter < max_iter and norm(prev_iteration_pagerank - next_iteration_pagerank, ord=1) > epsilon:

            y_hat = alpha * np.dot(adjacency_matrix, next_iteration_pagerank)
            y = y_hat + (1 - norm(y_hat, ord=1)) * v
            prev_iteration_pagerank, next_iteration_pagerank = next_iteration_pagerank, y / norm(y, ord=1)
            iter += 1
 
    return next_iteration_pagerank, iter
```

```python
rank, iter = page_rank(Q)
```

```python
dic = {"Url" : df.columns, "Rank" : rank}
ranked_DataFrame = pd.DataFrame(data=dic)
```

```python
ranked_DataFrame = ranked_DataFrame.sort_values(by="Rank", ascending=False).reset_index(drop=True)
```
